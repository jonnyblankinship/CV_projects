```python
import cv2 as cv 
import numpy as np
from IPython.display import display, clear_output 
from PIL import Image
from google.colab.patches import cv2_imshow

# Load two images
img1 = cv.imread('/content/drive/MyDrive/CS362V/Data/Logo.png')
img2 = cv.imread('/content/drive/MyDrive/CS362V/Data/Mustang.png')

#resize
w2 = int(img2.shape[1] * 0.5)
h2 = int(img2.shape[0] * 0.5)
dim = (w2, h2)
resized2 = cv.resize(img2, dim, interpolation = cv.INTER_AREA)

#make border
color = [255, 255, 255]
top, bottom = 563, 563
left, right = 374, 374
resized_with_border = cv.copyMakeBorder(resized2, top, bottom, left, right, cv.BORDER_CONSTANT, value=color)

#make roi
rows,cols,channels = img1.shape
roi = resized_with_border[0:rows, 0:cols]

#make grayscale
img2gray = cv.cvtColor(img1,cv.COLOR_BGR2GRAY)
cv2_imshow(img2gray)

#mask/inverse
ret,mask = cv.threshold(img2gray,200,255,cv.THRESH_BINARY)
cv2_imshow(mask)
mask_inv = cv.bitwise_not(mask)
cv2_imshow(mask_inv)

#add bg to mask
img2_bg = cv.bitwise_and(roi,roi,mask = mask)
cv2_imshow(img2_bg)

#add fg to mask inverse
img1_fg = cv.bitwise_and(img1,img1,mask = mask_inv)
cv2_imshow(img1_fg)

#combine fg and bg
final = cv.add(img2_bg,img1_fg)
cv2_imshow(final)
cv.imwrite('Homework1.png', final)

#count white pixels in img2_bg
rows2, cols2, channels2 = resized2.shape
roi2 = mask[563:1323, 374:1254]
ret,wp_thresh=cv.threshold(roi2,100,255,cv.THRESH_BINARY)
cv2_imshow(wp_thresh)
wp = np.sum(wp_thresh == 255)
print(wp)


###count blue pixels
img2_bw = cv.cvtColor(img2_bg,cv.COLOR_BGR2GRAY)
resized2_bw = cv.cvtColor(resized_with_border,cv.COLOR_BGR2GRAY)

#mask for original
ret,bp_or=cv.threshold(resized2_bw,70,255,cv.THRESH_BINARY_INV)
cv2_imshow(bp_or)

#mask for remaining
ret,bp_rem=cv.threshold(img2_bw,70,255,cv.THRESH_BINARY_INV)
cv2_imshow(bp_rem)

#show/count blue original
bluepix_or = cv.bitwise_and(resized_with_border,resized_with_border,mask = bp_or)
cv2_imshow(bluepix_or)
sumbp_or = np.sum(bluepix_or != 0)
print(sumbp_or)

#show/count blue remaining
bluepix_rem = cv.bitwise_and(img2_bg,img2_bg,mask = bp_rem)
cv2_imshow(bluepix_rem)
sumbp_rem = np.sum(bluepix_rem != 0)
print(sumbp_rem)


####count gold pixels
array_low = np.array([10,150,190])
array_high = np.array([140,250,252])
gp_rem=cv.inRange(img2_bg,array_low,array_high)
cv2_imshow(gp_rem)

#mask for original
gp_or = cv.inRange(resized_with_border,array_low,array_high)
cv2_imshow(gp_or)

goldpix_or = cv.bitwise_and(resized_with_border,resized_with_border,mask = gp_or)
cv2_imshow(goldpix_or)
gp_or = np.sum(goldpix_or != 0)
print(gp_or)

goldpix_rem = cv.bitwise_and(img2_bg,img2_bg,mask = gp_rem)
cv2_imshow(goldpix_rem)
gp_rem = np.sum(goldpix_rem != 0)
print(gp_rem)
```
